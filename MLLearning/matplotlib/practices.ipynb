```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
arr = np.linspace(0,10,100)
angle_x = np.sin(arr) # it takes value an arr as angle
angle_y = np.cos(arr) # it takes value an arr as angle
print("Sin Angle :",angle_x)
print("Cos Angle :",angle_y)
```

```python

plt.plot(arr,angle_x)
plt.title("Sin Wave")
plt.xlabel('Angle')
plt.ylabel('Sin values')
plt.show()
```

```python

plt.plot(arr,angle_y)
plt.title("Cos Wave")
plt.xlabel('Angle')
plt.ylabel('Cos values')
plt.show()
```

```python
# parabola curve
arr = np.linspace(-10,10,20)
parabola = arr**2
plt.plot(arr,parabola,'r*')
plt.title("Parabola Curve")
# plt.xlabel('Angle')
# plt.ylabel('Cos values')
plt.show()
```

```python
arr = np.linspace(12,20,30)
plt.title("Sin Cos Wave")
plt.plot(arr,np.sin(arr),'b-')
plt.plot(arr,np.cos(arr),'r--')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lang = ["Java","Python","Javascript","PHP","C#"]
peoples = [200,170,250,110,100]
ax.bar(lang,peoples)
plt.title("Developer in Domain")
plt.xlabel("Programming Languages")
plt.ylabel("Number of People")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lang = ["Java","Python","Javascript","PHP","C#"]
peoples = [200,170,250,110,100]
ax.pie(peoples,labels=lang,autopct='%1.1f%%')
plt.show()
```

```python
arr = np.linspace(0,10,30)
angle_x = np.sin(arr)
angle_y = np.cos(arr)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(arr,angle_x,color='g')
ax.scatter(arr,angle_y,color='b')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
arr = 20*np.random.random(100)
angle_x = np.sin(arr)
angle_y = np.cos(arr)
ax.scatter(arr,angle_x,angle_y,c=arr,cmap="Blues")
plt.show()
```
