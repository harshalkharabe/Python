```python
import seaborn as s
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Seaborn has some inbuilt dataset in it.
# total bill vs tip dataset
tips = s.load_dataset("tips")
```

```python
tips.head() # head() is used to give first 5 rows of the dataframe
```

```python
s.set_theme()
```

```python
s.relplot(data=tips,x="total_bill",y="tip",col='time',hue='smoker',style='smoker',size='size')
# s.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
iris = s.load_dataset('iris')
iris.head()
```

```python
s.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=iris)
```
