```python
"""line properties"""

# syntax : plt.plot(x,y,ls,lw,c)
# x : set of values on X-axis
# y : set of values on Y-axis
# ls : line style 
#    style          Description
# -----------------------------------
#     "-"           solid line
#     ":"           dotted line
#     "--"          dashed line
#     "-."          dasheddot line

# lw : line width in pixels
#  c : color of line


# Dwar a plot with following values
from matplotlib import pyplot as plt
x = [1,2,3,4]
y = [1,2,3,4]
plt.plot(x,y)
plt.show()

```

```python
# Marker properties
# 
# syntax : plt.plot(x,y,marker,mec,mfc,ms)
# x: set of values on X-axis
# y: set of values on Y-axis
# marker : 
#       marker_symbol       marker_value
# -------------------------------------------
#           'o'             circle
#           '*'             star
#           '.'             point
#           ',              pixel
# 
# mec : marker edge color
# mfc : marker face color
# ms : marker size

# Dwar a plot with following values
from matplotlib import pyplot as plt
x = [1,2,3,4]
y = [1,2,3,4]
plt.plot(x,y,marker='o',c="red")
plt.show()
```

```python
# Dwar a plot with following values
from matplotlib import pyplot as plt
x = [1,2,3,4]
y = [1,2,3,4]
# plt.plot(x,y,ls="dashdot")
plt.plot(x,y,marker="o",mec="r",mfc="yellow",ms=10)
plt.show()
```
