```python
# Marker properties
# 
# syntax : plt.plot(x,y,marker,mec,mfc,ms)
# x: set of values on X-axis
# y: set of values on Y-axis
# marker : 
#       marker_symbol       marker_value
# -------------------------------------------
#           'o'             circle
#           '*'             star
#           '.'             point
#           ',              pixel
# 
# mec : marker edge color
# mfc : marker face color
# ms : marker size

# Dwar a plot with following values
from matplotlib import pyplot as plt
x = [1,2,3,4]
y = [1,2,3,4]
plt.plot(x,y,marker='o',c="red")
plt.show()
```

```python
# setting the font properties
# Dwar a plot with following values
from matplotlib import pyplot as plt
x = [1,2,3,4]
y = [1,2,3,4]
plt.title("Graph",fontdict={"family":"times new roman","color":"r","size":16})
plt.xlabel("X-axis",fontdict={"family":"Arial","color":"b","size":15})
plt.ylabel("Y-axis",fontdict={"family":"serif","color":"b","size":15})
# plt.ylabel("Y-axis")
plt.plot(x,y,marker="x",mec="black",mfc="yellow",ms=10)
plt.show()
```

```python
# setting limites of X-axis and Y-axis
# xlim() and ylim()

# syntax : plt.xlim(lower_limit,max_limit)
# syntax : plt.ylim(lower_limit,max_limit)

# Dwar a plot with following values
from matplotlib import pyplot as plt
x = [1,2,3,4]
y = [7,2,3,4]
# plt.plot(x,y,ls="dashdot")
plt.xlim(1,10)
plt.ylim(1,10)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y,marker="x",mec="black",mfc="yellow",ms=10)
plt.show()
```

```python
# Drawing Multiple plots in same axes
# Dwar a plot with following values
from matplotlib import pyplot as plt
import numpy as np
x = np.array([1,2,3,4])
y = x**2
z = x**3
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,z,marker="*",mec="black",mfc="blue",ms=10)
plt.plot(x,y,marker="x",mec="black",mfc="yellow",ms=10)
plt.plot(x,x,marker="o",mec="red",mfc="red",ms=10)
plt.show()
```

```python
# legends and gride lines
# Drawing Multiple plots in same axes
# Dwar a plot with following values
from matplotlib import pyplot as plt
import numpy as np
x = np.array([1,2,3,4])
y = x**2
z = x**3
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
label = ["Green Flag","Red Flag","Blue Flag"]
plt.plot(x,z,marker="*",mec="black",mfc="blue",ms=10)
plt.plot(x,y,marker="x",mec="black",mfc="yellow",ms=10)
plt.grid(axis="both")
plt.plot(x,x,marker="o",mec="red",mfc="red",ms=10)
plt.legend(labels=label)
plt.show()
```
