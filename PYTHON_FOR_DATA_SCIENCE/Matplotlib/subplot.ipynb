```python
import pandas as pd
import matplotlib.pyplot as plt
d = {
    "Days":["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"],
    "Week1":[27,29,32,25,31,28,38],
    "Week2":[21,29,24,28,35,31,29],
    "Week3":[22,32,34,38,31,38,23]
}
df= pd.DataFrame(d)
plt.plot(df["Days"],df["Week1"],lw=4,c='r',marker="*",mec='black',ms=8)
plt.plot(df["Days"],df["Week2"],lw=4,c='b',marker="X",mec='black',ms=8)
plt.plot(df["Days"],df["Week3"],lw=4,c='g',marker="o",mec='black',ms=8)
plt.xlabel("Days")
plt.ylabel("Temprature")
plt.title("Weather")
plt.legend(labels=["Week1","Week2","Week3"])
plt.grid("both")
plt.show()
```
