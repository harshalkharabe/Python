```python
from matplotlib import pyplot as plt
```

```python
# syntax : plt.bar(x,y,color,width,labels) for vertical bar
# syntax : plt.barh(x,y,color,height,labels) for horizontal bar
```

```python
x=["Apple","Banana","Papaya","Watermelon"]
y=[23,25,29,27]
plt.bar(x,y)
plt.show()
```

```python
x=["Apple","Banana","Papaya","Watermelon"]
y=[23,25,29,27]
plt.bar(x,y,color='pink')
plt.show()
```

```python
x=["Apple","Banana","Papaya","Watermelon"]
y=[23,25,29,27]
col=["red","blue","green","yellow"]
plt.bar(x,y,color=col)
plt.show()
```

```python
x=["Apple","Banana","Papaya","Watermelon"]
y=[23,25,29,27]
col=["red","blue","green","yellow"]
plt.bar(x,y,color=col,label=["Apple","Banana","Papaya","Watermelon"])
# plt.legend(labels=["Apple","Banana","Papaya","Watermelon"])
plt.legend()
plt.show()
```

```python
x=["Apple","Banana","Papaya","Watermelon"]
y=[23,25,29,27]
col=["red","blue","green","yellow"]
plt.barh(x,y,color=col,label=["Apple","Banana","Papaya","Watermelon"])
# plt.legend(labels=["Apple","Banana","Papaya","Watermelon"])
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
d={
    "Savings":[20,30,40,50],
    "Emp":(102,103,104,105),
}
df = pd.DataFrame(d)
plt.bar(df["Savings"],df["Emp"],color=["Green","yellow"],width=5)
plt.legend()
plt.show()
```
