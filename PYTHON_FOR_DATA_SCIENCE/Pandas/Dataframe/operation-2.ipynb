```python
import pandas as pd

print("Adding data into dataframe")
df = pd.read_csv("std.csv")
print("Length of file :",len(df))
```

```python
print("Added new record into dataframe")
df.loc[len(df)+1]=[15,"Harsh K",39,89,67,99]
```

```python
print("After Adding new record :")
print(df)
```

```python
# Sorting data in dataframe
# syntax : Dataframe.sort_values(column_name)
df.sort_values(["maths"])
```

```python
# Sorting data of dataframe in reverse order
# syntax : Dataframe.sort_values(column_name)
# df.sort_values(["maths"])[::-1]
df.sort_values(["maths"],ascending=False)
```

```python
# Removing record or column from dataframe
# syntax : df.drop(column="column_name") for single column remove
# syntax : df.drop(column=["column_name","column_name"]) for multiple column remove
# syntax : df.drop(column="column_name",inplace=True) for single column remove
```

```python
import pandas as pd
# removing column temperory
df = pd.read_csv("std.csv")
print("Length of file :",len(df))
df.drop(columns="maths")

# for permanent change df.drop(columns="maths",inplace=True)
```

```python
import pandas as pd
# removing column temperory
df = pd.read_csv("std.csv")
print("Length of file :",len(df))
df.drop(columns=["maths","programming"])

# for permanent change df.drop(columns=["programming","maths"],inplace=True)
```

```python
# Removing records (rows)
# syntax : df.drop(labels=index_value,axis=0)
# syntax : df.drop(labels=[index_value,index_value1],axis=0)
print(df)
print(len(df))
print('-'*80)
# removing single value
print(df.drop(labels=3))
print(len(df.drop(labels=2)))
```

```python
# Removing multiple records (rows)
# syntax : df.drop(labels=index_value,axis=0)
# syntax : df.drop(labels=[index_value,index_value1],axis=0)
print(df)
print(len(df))
print('-'*80)
# removing single value
print(df.drop(labels=[3,6,8]))
print(len(df.drop(labels=[3,6,8])))
```
