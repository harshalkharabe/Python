```python
import pandas as pd
df = pd.read_csv("std.csv")
print(df)
```

```python
# Adding new column
# syntax : df["New_column_name"] = Defalut_value
# syntax : df["New_column_name"] = Expression

# Adding new column Total marks
# using default value
# df["Total"]=0
```

```python
# Adding new column into df
# using expression
print("After Adding new column :")
df["Total"]=df["english"]+df["hindi"]+df["maths"]+df["programming"]
df
```

```python
print("After adding Percentage Column :")
df["Percentage"]=round((df["Total"]/400)*100,2)
df
```

```python
print("Maximum Percentage student :")
df["Percentage"].max()
```

```python
# Data Filtering
# get the students who's marks is greater than 90
df[df["maths"]>=90]["Name"]
```

```python
print("Filtering on Single Condition")
df.loc[df['maths']<90,['Name','Total','Percentage']]
```

```python
print("Multiple conditions")
df.loc[(df['maths']>=86) & (df['maths']<91),['Name','Total','Percentage']]
```

```python
# Give the grade to the students with following :
# A grade --> greater than 95
# B grade less than 75 and greater than 60
# C grade less than 60
df["Grade"]=None

print("A grade students :")
df.loc[df['Percentage']>=90,['Grade']]="A"
df.loc[(df['Percentage']<90) & (df['Percentage']>=83),['Grade']]="B"
df.loc[df['Percentage']<83,['Grade']]="C"
print(df)
```

```python
print("Total number of people having A grade",df[df["Grade"]=="A"].shape[0])
```

```python
# Find the name of of student who got heighest percentage
print("Topper student :")
df.loc[df["Percentage"].max()==df["Percentage"]]
```
